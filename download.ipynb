```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("raw data.xlsx")
```

```python
df.head()
df.info()
df.describe()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df['Sale_Date'] = pd.to_datetime(df['Sale_Date'], errors='coerce')
```

```python
df['Customer_Type'] = df['Customer_Type'].str.strip().str.title()
df['Payment_Method'] = df['Payment_Method'].str.strip().str.title()
df['Region'] = df['Region'].str.strip().str.title()
df['Sales_Rep'] = df['Sales_Rep'].str.strip().str.title()
```

```python
df['Profit'] = (df['Unit_Price'] - df['Unit_Cost']) * df['Quantity_Sold']
```

```python
df['Discount_Amount'] = df['Sales_Amount'] * df['Discount']
```

```python
df.to_csv("raw data.csv", index=False)
```
